```python
#Foi usado a vizinhança 4
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image 


def dfs(img, colunas, linhas):
    cont  = 1
    for i in range(colunas):
        for j in range(linhas):
            if(img[i, j]!=1):
                dfs_visita(img, i,j,cont)
                cont=cont+1;
    print("Quantidade de componentes: ",cont-1);

def dfs_visita(img, i, j, cont):
    pilha = [];
    pilha.append((i,j))
    while(len(pilha)>0):
        coord = pilha.pop()
        y = coord[0]
        x = coord[1]
        img[y,x]=1
        imgRot[y,x] = cont
        if(img[y-1,x]!=1):
            pilha.append((y-1,x));
        if(img[y,x+1]!=1):
            pilha.append((y, x+1));
        if(img[y+1,x]!=1):
            pilha.append((y+1, x));
        if(img[y,x-1]!=1):
            pilha.append((y, x-1));


img = cv2.imread("art8.png", cv2.IMREAD_GRAYSCALE)

imgRot = np.zeros((img.shape[0], img.shape[1]))

dfs(img, img.shape[0], img.shape[1])

imgFinal = 255 * (imgRot - imgRot.min()) // (imgRot.max() - imgRot.min())

imgFinal = Image.fromarray(imgFinal)

plt.imshow(imgFinal); 
```
