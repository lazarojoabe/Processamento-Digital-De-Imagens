1)	Abrir as imagens coloridas e mostrar a imagem e seus histogramas (separados) cada canal de cor. Considerar os seguintes sistemas de cores:
a.	RGB
b.	HSV ou HSI
c.	Lab


```python
def calcHist(img):
    histograma_verm = np.zeros(256)
    histograma_verde = np.zeros(256)
    histograma_azul = np.zeros(256)

    # Obter as dimenses da imagem
    altura, largura, _ = img.shape
    for i in range(altura):
        for j in range(largura):
            pixel_value = img[i, j]
            histograma_verm[pixel_value[0]] += 1
            histograma_verde[pixel_value[1]] += 1
            histograma_azul[pixel_value[2]] += 1
    return histograma_verm, histograma_verde, histograma_azul

img1 = cv2.imread("image.jpg", cv2.IMREAD_COLOR)
img1RGB = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread("image2.jpg", cv2.IMREAD_COLOR)
img2RGB = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img3 = cv2.imread("Lenna.jpg", cv2.IMREAD_COLOR)
img3RGB = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

img1HSV = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
img2HSV = cv2.cvtColor(img2, cv2.COLOR_BGR2HSV)
img3HSV = cv2.cvtColor(img3, cv2.COLOR_BGR2HSV)

hist1 = calcHist(img1RGB)
hist2 = calcHist(img2RGB)
hist3 = calcHist(img3RGB)

fig = plt.figure( figsize=(24, 8))

plt.subplot(2, 3, 1).set_title("Imagem 1")
plt.bar(range(256), hist1[0], alpha=0.75, color='r', label='Histograma-Vermelho')
plt.axis( [0, 255, 0, 12000] )


plt.bar(range(256), hist1[1], alpha=0.75, color='g', label='Histograma-Verde')
plt.axis( [0, 255, 0, 12000] )

plt.bar(range(256), hist1[2], alpha=0.75, color='b', label='Histograma-Azul');
plt.axis( [0, 255, 0, 12000] )

plt.subplot(2, 3, 2).set_title("Imagem 2")

plt.bar(range(256), hist2[0], alpha=0.75, color='r', label='Histograma-Vermelho')
plt.axis( [0, 255, 0, 17000] )

plt.bar(range(256), hist2[1], alpha=0.75, color='g', label='Histograma-Verde')
plt.axis( [0, 255, 0, 17000] )

plt.bar(range(256), hist2[2], alpha=0.75, color='b', label='Histograma-Azul');
plt.axis( [0, 255, 0, 17000] )

plt.subplot(2, 3, 3).set_title("Imagem 3")

plt.bar(range(256), hist3[0], alpha=0.75, color='r', label='Histograma-Vermelho')
plt.axis( [0, 255, 0, 10000] )

plt.bar(range(256), hist3[1], alpha=0.75, color='g', label='Histograma-Verde')
plt.axis( [0, 255, 0, 10000] )

plt.bar(range(256), hist3[2], alpha=0.75, color='b', label='Histograma-Azul');
plt.axis( [0, 255, 0, 10000] )

plt.subplot(2, 3, 4)
plt.imshow(img1RGB)

plt.subplot(2, 3, 5)
plt.imshow(img2RGB)

plt.subplot(2, 3, 6)
plt.imshow(img3RGB)

plt.show()
```

2)	Utilizando a imagem do mandrill, faa rotinas para deteco do focinho (regio vermelha e azul). Utilize algum algoritmo de limiarizao.

```python
def calcVarB_e_VarG(k, histograma):
    P1 = np.sum(histograma[:k])
    P2 = 1 - P1

    m1 = np.sum(np.arange(k) * histograma[:k]) / (P1 + 1e-10)  # Evitar diviso por zero
    m2 = np.sum(np.arange(k, 256) * histograma[k:]) / (P2 + 1e-10)  # Evitar diviso por zero
    
    mG = P1 * m1 + P2 * m2
    varB = P1 * P2 * (m1 - m2)**2

    varG = np.sum(histograma * (np.arange(256) - mG)**2)

    return varB, varG

def eta(k, histograma):
    result = calcVarB_e_VarG(k, histograma)
    varB = result[0]
    varG = result[1]
    return varB / (varG + 1e-10)  # Evitar diviso por zero

def acharMelhorKEMaiorEta(histograma):
    maiorEta = 0
    melhorK = 0
    for i in range(0, 256):
        valor = eta(i, histograma)
        if maiorEta < valor:
            melhorK = i
            maiorEta = valor
    return melhorK, maiorEta



img_mand = cv2.imread("mandrill.tiff", cv2.IMREAD_COLOR)
img_mand_RGB = cv2.cvtColor(img_mand, cv2.COLOR_BGR2RGB)


plt.imshow(img_mand_RGB);
```

3)	Utilizando a imagem HE.jpg, tente detectar e contar a quantidade de ncleos celulares (em azul). Aplicar o Ostu para os 9 canais de cores e avaliar qual foi melhor (R,G,B,H,S,V,L,a,b)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('HE.jpg')

# Convertendo a imagem para o espao de cor HSV (Hue, Saturation, Value)
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_blue = np.array([100, 100, 100], dtype=np.uint8)
upper_blue = np.array([140, 255, 255], dtype=np.uint8)

# Criando uma mscara para os tons de azul
blue_mask = cv2.inRange(hsv_img, lower_blue, upper_blue)

# Encontrando contornos na imagem segmentada
contours, _ = cv2.findContours(blue_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Desenhando os contornos na imagem original
img_contours = img.copy()
cv2.drawContours(img_contours, contours, -1, (0, 255, 0), 2)

# Contando os ncleos celulares
num_nucleos = len(contours)

# Mostrando as imagens resultantes
titles = ['Original Image', 'Blue Nuclei Segmentation', 'Contours']
images = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB), blue_mask, cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB)]

img_contours_rgb = cv2.cvtColor(img_contours, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(24, 8))

plt.subplot(1, 2, 1).set_title("Imagem Original:")
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2).set_title("Imagem com ncleos azuis detectados:")
plt.imshow(img_contours_rgb)
plt.axis('off')

plt.show()

print(f'Nmero de ncleos azuis na imagem: {num_nucleos}')
# Converter img_contours para RGB antes de salvar
img_contours_rgb_for_saving = cv2.cvtColor(img_contours_rgb, cv2.COLOR_BGR2RGB)

output_path = 'imagem_com_contornos.jpg'
cv2.imwrite(output_path, img_contours_rgb_for_saving)

print(f'Imagem com contornos salva em: {output_path}')
```
