1) Aplique nas imagens abaixo o algortimo Watershed para realizar a contagem de células. Faça tanto para  a imagem original e para a imagem gradiente

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#li a imagem como colorida para poder usar na cv2.watershed, porque ela não aceita imagens em escala de cinza
img = cv2.imread("image.jpg", cv2.IMREAD_COLOR)  # Alterado para IMREAD_COLOR

# Converter a imagem para escala de cinza
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Thresholding da imagem
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

kernel = np.ones((3, 3), np.uint8)

opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=10)

# Background
background = cv2.dilate(opening, kernel, iterations=10)

# Foreground
dist = cv2.distanceTransform(opening, cv2.DIST_L2, 5)

_, foreground = cv2.threshold(dist, 0, 255, cv2.THRESH_BINARY)

foreground = np.uint8(foreground)

# Pixels desconhecidos
unknown = background - foreground

_, markers = cv2.connectedComponents(foreground)
#corr
markers = markers + 1
markers[unknown == 255] = 0

newImg = cv2.watershed(img, markers)

# Corrigindo a atribuição de valores em img
img[markers == -1] = [255, 0, 0]  # Atribuindo valores de cor vermelha

print("QTD: ", len(markers))

plt.imshow(newImg)
plt.show()
```
